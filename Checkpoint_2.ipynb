### Data Cleaning
Before we can move on to plotting, we need to strip out the numeric figures for the some of the features that are currently viewed as strings/objects by pandas. For the sake of simplicity, and because we will be discussing data processing in depth later on, we will keep the pre-processing steps minimal in order to yield something that we can use to plot the data. This will mainly take the form of isolating the numerical values from the column entries. For now, we will be splitting the courses into their department, number, and name, the quarter into its quarter name and year, grabbing the percentages in % rec class and % rec prof, and extracting the float values from avg GPA received/expected. We will also consider normalizing our input data later to make training the ANN faster, but for the purposes of plotting right now, we want to keep the values and units in their natural form for greater interpretability.

```python
# Splitting the Course column
df['Department'] = df['Course'].str.extract(r'(\w+)\s')
df['Course Number'] = df['Course'].str.extract(r'(\w+\s\d+)')
df['Course Name'] = df['Course'].str.extract(r'-\s(.*)\s\(')

# Extracting Quarter Abbreviation and Year
df['Quarter Abbreviation'] = df['Quarter'].str.extract(r'([A-Z]+)')
df['Year'] = "20" + df['Quarter'].str.extract(r'(\d+)$')

# Converting percentages to decimals
df['Proportion Recommended Class'] = df['Percentage Recommended Class'].str.rstrip('%').astype('float') / 100
df['Proportion Recommended Professor'] = df['Percentage Recommended Professor'].str.rstrip('%').astype('float') / 100

# Extracting numeric values from grades
df['Average GPA Expected'] = df['Average Grade Expected'].str.extract(r'\((\d+\.\d+)\)')[0].astype('float')
df['Average GPA Received'] = df['Average Grade Received'].str.extract(r'\((\d+\.\d+)\)')[0].astype('float')
```

### Data Usage
It is important to note that this dataset does not contain information about all classes that UCSD has to offer. This is because there are some classes that do not participate in the CAPEs process (MCWP classes) and thus did not show up on the CAPEs website and was subsequently not scraped and added to the dataset when it was created. Additionally, there are some classes where the number of capes given were very low relative to the number of people taking the class. We've noticed this is usually the case with classes that are offered during the Summer where the number enrolled is usually lower. This may present issues as the values present in other features may be highly influenced by the voices of a few, which may not reflect the true general perception of a class. Thus, it may be appropriate to exclude entries where the ratio of CAPEs given to the number enrolled is too low (below a certain threshold). Additionally, we should note that each observation in the dataset also comes with an evaluation link, which contains additional information about a class (ex. how much students felt the exams represented the course content). This is a treasure trove of extra information that we can add to our dataset if we later feel it is necessary. To do this, we'd have to write our own webscraping algorithm and run it on all the observations in the database, which will probably take a while and will need to be verified for correctness.

It looks most of our data is not a number, which means we can't do much computation until we fix that. Let's start with the easy ones first.

* It seems that % rec class and % rec prof are basically numbers, but since there's a % appended to them, pandas thinks they're strings/objects. We can fix those by just replacing each of these values with their numerical equivalent (i.e. 50% = 0.5).
* Average grade expected/received are also bascically numbers, but since there's a letter grade prepended to them, pandas also thinks they're strings/objects. We can easily fix this by just stripping the letter grade and parentheses for all observations with a value for that feature.




```python
# Processing % rec class/prof

# just take off the % and convert resulting string to float, then divide by 100
pp_df['Percentage Recommended Class'] = pp_df['Percentage Recommended Class'].str.rstrip('%').astype(float)/100
pp_df['Percentage Recommended Professor'] = pp_df['Percentage Recommended Professor'].str.rstrip('%').astype(float)/100
```

```python
pp_df.dtypes
```

```python
# Processing average grade received/expected

# Use RegEx to extract just the numbers
pp_df['Average Grade Expected'] = pp_df['Average Grade Expected'].str.extract(r'(\d+\.\d+)').astype(float)
pp_df['Average Grade Received'] = pp_df['Average Grade Received'].str.extract(r'(\d+\.\d+)').astype(float)
```

The remaining features don't have any obvious ways of converting the observations into numerical values, but we can figure out a way to do so by encoding the values into numbers. Let's tackle course and quarter since they at least have some numbers in them.
* We can try to apply one hot encoding to quarter by adding in a feature vector for each quarter a class can be offered. This is fine, since there are only 6 categories we need to consider (Fall, Winter, Spring, SS1, SS2). To take care of the year, we can also extract the year a class was offered and make that its own observation in a new feature column 'year'

* For courses, since they're all formatted as DEPT # - name, we could try to isolate the DEP and # since that's all that's needed to identify a class. Since there are so many departments, we'd want to use label encoding, and we can do so by just assigning a number to a department name based on how they rank alphabetically. Additionally, since the class number is basically, a number, we can just directly convert it into an int.



```python
# let's first find out what the possible quarters are

#print(df['Quarter'].shape)

# isolate the years
years = pp_df['Quarter'].str[2:4].astype(int)

# isolate the quarter
quarters = pp_df['Quarter'].str[:2]
```

```python
print(years.unique())
print(quarters.unique())
```

```python
# what is S3 and SU?
```

```python
# make year 20xx
years_20XX = years + 2000

# append years to the array
pp_df['Year'] = years_20XX

pp_df.head()

# replace quarter with just label encoded value

pp_df['Quarter'] = quarters

quarter_map = {'FA': 0, 'WI': 1, 'SP': 2, 'S1': 3, 'S2': 4, 'S3': 5, 'SU': 6}

pp_df['Quarter'] = pp_df['Quarter'].replace(quarter_map)

pp_df.head()
```

```python
# now let's fix the course feature
department = df['Course'].str.extract(r'([A-Za-z\s]+)\d+')
print(department[0].unique())

course_nums = df['Course'].str.extract(r'(\d+)').astype(int)
print(course_nums[0].unique())
```

```python
pp_df['Department'] = department
pp_df['Course Number'] = course_nums

pp_df.head()
```

```python
# create a mapping for department
import numpy as np

dept_sorted = np.sort(department[0].unique())

print(dept_sorted)

dept_map = {}
counter = 0
for i in range(len(dept_sorted)):
  if (dept_map.get(dept_sorted[i]) == None):
    dept_map[dept_sorted[i]] = counter
    counter += 1

pp_df['Department'] = pp_df['Department'].replace(dept_map)
```

```python
# since we have just encoded all the info from the course column, we can safely drop it from our dataset
pp_df = pp_df.drop('Course', axis = 1)
pp_df.head()
```

Now we need to deal with the instructor and eval URL columns. Since eval URL doesn't contain any useful information for us (unless we want to use it to get more data), we can safely drop it. Additionally, for ethical and privacy reasons we think we should remove any association of professors to classes for the data set. This is because if this project ends up being used to help UCSD plan their courses, we wouldn't want the model to associate success with a particular professor, which might lead to workplace issues (i.e. favoritism). We already have a metric of measuring how much students enjoy their professors (Percentage Recommended Professor) so the names are not needed, and it's better they stay anonymous too for the reason stated above.



```python
pp_df = pp_df.drop('Instructor', axis = 1)
pp_df = pp_df.drop('Evalulation URL', axis = 1)
pp_df
```
