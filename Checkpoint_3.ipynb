It looks most of our data is not a number, which means we can't do much computation until we fix that. Let's start with the easy ones first.

* It seems that % rec class and % rec prof are basically numbers, but since there's a % appended to them, pandas thinks they're strings/objects. We can fix those by just replacing each of these values with their numerical equivalent (i.e. 50% = 0.5).
* Average grade expected/received are also bascically numbers, but since there's a letter grade prepended to them, pandas also thinks they're strings/objects. We can easily fix this by just stripping the letter grade and parentheses for all observations with a value for that feature.




```python
# Processing % rec class/prof

# just take off the % and convert resulting string to float, then divide by 100
pp_df['Percentage Recommended Class'] = pp_df['Percentage Recommended Class'].str.rstrip('%').astype(float)/100
pp_df['Percentage Recommended Professor'] = pp_df['Percentage Recommended Professor'].str.rstrip('%').astype(float)/100
```

## Imputation

```python
na_counts = pp_df.isna().sum()
na_counts
```

We have NaN values for Study Hours per Week, Average Grade Expected, and Average Grade Received. We will go through the process of imputation to fill in the NaN values. For the one NaN in Study Hours Per Week, we will look at all other iterations of the course and find the mean Study Hours Per Week. For Average Grade Expected and Average Grade Received, we will look at all other iterations of the course that contains the NaN value and fill it in with the average.

```python
# Filling the NaN for Study Hours Per Week

study_hours_nan_row = pp_df[pp_df['Study Hours per Week'].isna()]

dept = study_hours_nan_row['Department'].values[0]
course_num = study_hours_nan_row['Course Number'].values[0]
course_suff = study_hours_nan_row['Course Suffix'].values[0]

averages = pp_df.groupby(['Department', 'Course Number', 'Course Suffix'])['Study Hours per Week'].transform('mean')

pp_df['Study Hours per Week'] = pp_df['Study Hours per Week'].fillna(averages)
```

```python
# Filling the NaN for Average Grade Expected

expected_nan_row = pp_df[pp_df['Average Grade Expected'].isna()]

dept = expected_nan_row['Department'].values[0]
course_num = expected_nan_row['Course Number'].values[0]
course_suff = expected_nan_row['Course Suffix'].values[0]

averages = pp_df.groupby(['Department', 'Course Number', 'Course Suffix'])['Average Grade Expected'].transform('mean')

pp_df['Average Grade Expected'] = pp_df['Average Grade Expected'].fillna(averages)
```

```python
# Filling the NaN for Average Grade Received

received_nan_row = pp_df[pp_df['Average Grade Received'].isna()]

dept = received_nan_row['Department'].values[0]
course_num = received_nan_row['Course Number'].values[0]
course_suff = received_nan_row['Course Suffix'].values[0]

averages = pp_df.groupby(['Department', 'Course Number', 'Course Suffix'])['Average Grade Received'].transform('mean')

pp_df['Average Grade Received'] = pp_df['Average Grade Received'].fillna(averages)

pp_df.isna().sum()
```

After imputing the NaN values to be the means of all other iterations of the course, we still observe that we have NaN values for Average Grade Expected and Average Grade Received... Let's investigate why this is the case!

```python
# Why do we still have NaN values for Average Grade Expected?

na_rows = pp_df[pp_df['Average Grade Expected'].isna()]
num_rows = pp_df[pp_df['Average Grade Expected'].isna()].shape[0]

count = 0

for i in range(0, num_rows):
  row = na_rows.iloc[i]

  dept = row['Department']
  course_num = row['Course Number']
  course_suff = row['Course Suffix']

  filtered_rows = pp_df[(pp_df['Department'] == dept) &
                        (pp_df['Course Number'] == course_num) &
                        (pp_df['Course Suffix'] == course_suff)]

  if (len(filtered_rows) == 1):
    count += 1
  elif (len(filtered_rows) == filtered_rows['Average Grade Expected'].isna().sum()):
    count += 1
  else:
    continue

print(count)
print(pp_df['Average Grade Expected'].isna().sum())
```

As I suspected! My initial thought as to why there were still NaNs remaining was that these remaining classes only have **one** iteration of the course in the data set *or* that **all** iterations of the course in the data set have the Average Grade Expected as NaN. For example, if I have only one iteration of a class and it has a NaN value, I cannot impute the mean from other iterations because there is only one -- the one with the NaN. Likewise, if I try to impute for a NaN value in a course where *all* its iterations have NaN values, then I cannot impute a value there because there is no legitimate value to obtain a mean from.

<br>

In the code above, I filtered the data frame to include the observations that had the courses that I observed had a NaN for Average Grade Expected, trying to look for all iterations of the course. Then, I checked the filtered data frame. If there was only one row in the filtered data frame, then that meant there was only one iteration of the course and it had a NaN, so we cannot impute a mean here. If there was more than one row in the filtered data frame, then we check if each row has a NaN for average grade expected. If that is the case, then that means we also cannot impute a mean since all iterations have a NaN value. I incremented a count that tracked the instances where we could not impute and I compared it to the number of NaN observations for Average Grade Expected and they were equal! So that means that all the remaining NaN values for Average Grade Expected could not be imputed using the average for the other iterations of the course!

<br>

For Average Grade Received, we will follow the same method below...

```python
# Why do we still have NaN values for Average Grade Received?

na_rows = pp_df[pp_df['Average Grade Received'].isna()]
num_rows = pp_df[pp_df['Average Grade Received'].isna()].shape[0]

count = 0

for i in range(0, num_rows):
  row = na_rows.iloc[i]

  dept = row['Department']
  course_num = row['Course Number']
  course_suff = row['Course Suffix']

  filtered_rows = pp_df[(pp_df['Department'] == dept) &
                        (pp_df['Course Number'] == course_num) &
                        (pp_df['Course Suffix'] == course_suff)]

  if (len(filtered_rows) == 1):
    count += 1
  elif (len(filtered_rows) == filtered_rows['Average Grade Received'].isna().sum()):
    count += 1
  else:
    continue

print(count)
print(pp_df['Average Grade Received'].isna().sum())
```

Same result! All the remaining observations where the Average Grade Received is a NaN cannot be imputed based off the mean of all other iterations of the course for reasons mentioned previously.

<br>

**How will we solve this and impute the remaining NaNs?**
<br>
A solution is instead of looking at the iterations of the specific course and imputing the mean, we can broaden our scope and instead look at the mean of all courses in the general department. For example, say we have a NaN value for the course CSE 151A and we cannot impute a mean from other iterations because there is either only one iteration that has a NaN or all iterations of the course have a NaN... then, what we can do is impute the mean from all courses in the CSE department.

```python
# Filling in the remaining NaNs for Average Grade Expected

expected_by_dept = pp_df.groupby('Department')['Average Grade Expected'].transform('mean')

pp_df['Average Grade Expected'] = pp_df['Average Grade Expected'].fillna(expected_by_dept)

received_by_dept = pp_df.groupby('Department')['Average Grade Received'].transform('mean')

pp_df['Average Grade Received'] = pp_df['Average Grade Received'].fillna(received_by_dept)

print(pp_df.isna().sum())
```

Looks like we managed to significantly reduce the number of NaNs from 84 to 1 for Average Grade Expected and from 5989 to 342 for Average Grade Received. But **why are we still getting NaNs?**

<br>

Let's investigate!

```python
# Looking at remaining NaNs for Average Grade Expected

na_rows = pp_df[pp_df['Average Grade Expected'].isna()]
num_rows = pp_df[pp_df['Average Grade Expected'].isna()].shape[0]

count = 0

for i in range(0, num_rows):
  row = na_rows.iloc[i]

  dept = row['Department']

  filtered_rows = pp_df[(pp_df['Department'] == dept)]

  if (len(filtered_rows) == 1):
    count += 1
  elif (len(filtered_rows) == filtered_rows['Average Grade Expected'].isna().sum()):
    count += 1
  else:
    continue

print(count)
print(pp_df['Average Grade Expected'].isna().sum())
```

```python
# Looking at remaining NaNs for Average Grade Expected

na_rows = pp_df[pp_df['Average Grade Received'].isna()]
num_rows = pp_df[pp_df['Average Grade Received'].isna()].shape[0]

count = 0

for i in range(0, num_rows):
  row = na_rows.iloc[i]

  dept = row['Department']

  filtered_rows = pp_df[(pp_df['Department'] == dept)]

  if (len(filtered_rows) == 1):
    count += 1
  elif (len(filtered_rows) == filtered_rows['Average Grade Received'].isna().sum()):
    count += 1
  else:
    continue

print(count)
print(pp_df['Average Grade Received'].isna().sum())
```

After further investigation, it looks like the reason why there are still NaNs is the same exact reasoning as when we tried to impute based on Department, Course Number, and Course Suffix but this time only for Department. There is only one observation with the specified department and it is NaN *or* there are multiple observations with the specified department and they are all NaN.

<br>

So again, **how will we solve this?**
<br>

Well, let's go even broader... Instead of imputing based on the mean of the courses in the department, we will impute based on the mean of the entire data set. And this last resort should get rid of the NaNs once and for all!

```python
# NaNs be gone!

mean_expected = pp_df['Average Grade Expected'].mean()

pp_df['Average Grade Expected'] = pp_df['Average Grade Expected'].fillna(mean_expected)

mean_received = pp_df['Average Grade Received'].mean()

pp_df['Average Grade Received'] = pp_df['Average Grade Received'].fillna(mean_received)

pp_df.isna().sum()
```

Hooray!

### Training the Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import RMSprop
```

```python
target_col = "Average Grade Received"
train_cols = ["Total Enrolled in Course",
              "Percentage Recommended Class",
              "Percentage Recommended Professor",
              "Study Hours per Week",
              "Average Grade Expected"
              ]

X = pp_df[train_cols]
y = pp_df[target_col]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
print(X_train.head(), '\n')
print(y_train.head())
```

```python
# Build the model
model_1 = Sequential([
    Dense(units = 128, activation='linear', input_dim = X_train.shape[1]), # input layer + hidden layer 1
    Dense(units = 64, activation='tanh'), # hidden layer 2
    Dense(units = 32, activation='relu'), # hidden layer 3
    Dense(units = 16, activation='relu'), # hidden layer 4
    Dense(units = 8, activation='relu'), # hidden layer 5
    Dense(units = 4, activation='relu'), # hidden layer 6
    Dense(units = 2, activation='relu'), # hidden layer 7
    Dense(units = 1) # no activation for output layer as we want a continuous value
])

# Compile the model
model_1.compile(optimizer='adam', loss='mse')
```

```python
# Set up early stoppping
early_stopping = tf.keras.callbacks.EarlyStopping(
    #Enter your parameters
    monitor='loss',
    min_delta=0,
    patience=5,
    verbose=0,
    mode='auto',
    baseline=None,
    restore_best_weights=False, # disable since we are using checkpointing
    start_from_epoch=0

)

# Set up model checkpoint
model_checkpoint = tf.keras.callbacks.ModelCheckpoint(
    #Enter your paramaters
    filepath = '/content/checkpoint_v1',
    save_weights_only=True,
    monitor='val_loss', # save the weights that results in the lowest evaluation loss
    mode='auto',
    save_best_only=True
)
```

```python
# Train the model
history = model_1.fit(X_train, y_train, batch_size = 10, epochs=30, validation_split=0.3, callbacks = [early_stopping, model_checkpoint], verbose=1)
```

```python
# load the weights that resulted in the lowest val_loss
model_1.load_weights('/content/checkpoint_v1')
```

### Evaluating the Model

```python
# evaluate training vs test
yhat_train = model_1.predict(X_train)
yhat_test = model_1.predict(X_test)
mse_train = mean_squared_error(y_train, yhat_train)
mse_test = mean_squared_error(y_test, yhat_test)

print("MSE Train: {}".format(mse_train))
print("MSE Test: {}".format(mse_test))
```

As you can see, it looks like our model performs pretty similarly when presented with unseen test data as it does with the data it was trained on. In fact, the results seem to indicate that the model performs slightly better with the provided test data than the training data as the test MSE is slightly lower than the training MSE. We belive this is probably just due to luck (the seed that we used to split the data) and that if we performed k cross fold validation we may see slightly different results. Ultimately, though the loss is not perfect (MSE is not 0) for either dataset, the loss values that we do get are rather low, which suggests that our model would do a good job at preducting the average GPA received for a class given the values for "Total Enrolled in Course", "Percentage Recommended Professor", "Study Hours per Week", and "Average Grade Expected". These are quite promising results for our first model!
